```python
from ultralytics import YOLO
```

```python
model = YOLO("/content/runs/detect/train/weights/last.pt")
```

```python
import cv2
import numpy as np
```

```python
cap = cv2.VideoCapture(0)

while cap.isOpened():
  ret,frame = cap.read()

  result = model(frame)

  cv2.imshow("YOLO",np.squeeze(result.render()))
  if cv2.waitKey(10) & 0xFF == ord("c"):
        break

cap.release()
cv2.destroyAllWindows()

  
```
